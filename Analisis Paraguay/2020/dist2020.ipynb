```python
mes = ['enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio',
      'julio', 'agosto', 'setiembre', 'octubre', 'noviembre', 'diciembre']
mes_code = dict(zip(mes,['01','02','03','04','05','06',
                     '07','08', '09','10','11','12']))
dia = []
for f in range(1,32):
    dia.append(f)
path_init = os.path.join('data', 'earthpy-downloads', 'rayos-paraguay',
               'dmh', '2020'+'\\')
list1 = [path_init + f for f in os.listdir(path_init) if f.startswith('2020_')]
list2 = []
for i in range(len(list1)):
    list2.append([list1[i]+ '\\' +
                  f for f in os.listdir(list1[i]) if f.startswith('paraguay_')])

list2 = dict(zip(mes[2:],list2))
```

```python
desc = {}
for i in list2.keys():
    aux = []
    print(i)
    for j in range(len(list2[i])):
        print('\t',j)
        try:
            aux.append(pd.read_csv(list2[i][j], parse_dates=['LightningTime'],index_col=['LightningTime'],
                               delimiter = ',', engine = 'python', on_bad_lines = 'warn'))
        except:
            print('\t\tNo data update')
            continue
    desc[i] = aux
```

```python
for i in desc.keys():
    for j in desc[i]:
        j.index = pd.to_datetime(j.index, infer_datetime_format= True)
        coord = gpd.points_from_xy(j.Longitude, j.Latitude)
        j = gpd.GeoDataFrame(data = j, geometry = coord, crs = pais.crs)
        j = gpd.clip(j, pais)
```

```python
desc20 = {}
for i in desc.keys():
    print(i)
    aux = []
    for j in range(1, dia_mes['n'][i]+1):
        print(j)
        df = gpd.GeoDataFrame()
        for k in desc[i]:
            df = pd.concat([df, k[(k.index.month == int(mes_code[i])) & (k.index.day == j)]],
                          axis = 0)
        aux.append(df)
    desc20[i] = dict(zip(dia,aux))
```
