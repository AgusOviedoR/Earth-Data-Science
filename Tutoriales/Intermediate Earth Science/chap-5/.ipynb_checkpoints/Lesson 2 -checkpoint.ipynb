```python
#Importando librerias
import os

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import rioxarray as rxr
import earthpy as et

#Configuracion de grafico
sns.set(font_scale = 1.5, style = 'whitegrid')

os.chdir(os.path.join(et.io.HOME, 'earth-analytics', 'data'))

#Datos
lidar_dem_path = os.path.join('earthpy-downloads','colorado-flood','spatial',
                             'boulder-leehill-rd', 'pre-flood', 'lidar',
                             'pre_DTM.tif')
lidar_dem_xr = rxr.open_rasterio(lidar_dem_path, masked=True).squeeze()
lidar_dem_xr
```

```python
lidar_dsm_path = os.path.join('earthpy-downloads', 'colorado-flood', 'spatial',
                             'boulder-leehill-rd','pre-flood', 'lidar','pre_DSM.tif')
lidar_dsm_xr = rxr.open_rasterio(lidar_dsm_path, masked=True).squeeze()
lidar_dsm_xr
```

```python
#Necesitamos saber si los limites de ambos archivos son los mismos,
#tambien la resolucion debe ser la misma
print(lidar_dem_xr.rio.bounds() == lidar_dsm_xr.rio.bounds())
print(lidar_dem_xr.rio.resolution() == lidar_dsm_xr.rio.resolution())
```

```python
#Canopy Height Model = DSM - DTM
lidar_chm_xr = lidar_dsm_xr - lidar_dem_xr

#Grafico
f, ax = plt.subplots(figsize = (10,5))
lidar_chm_xr.plot(cmap = 'Greens')
ax.set(title = 'Canopy Height Model for Lee Hill Road Pre-Flood')
ax.set_axis_off()
plt.show()
```

```python
lidar_chm_xr.plot.hist()
plt.show()
```

```python
#Guardando el archivo de chm calculado
data_path = os.path.join('earthpy-downloads', 'colorado-flood',
                        'spatial','outputs')
if os.path.exists(data_path):
    print('The directory ', data_path, 'exists!')
else:
    os.makedirs(data_path)
```

```python
pre_chm_data_path = os.path.join(data_path, 'pre-flood-chm.tif')
pre_chm_data_path

#Exportando datos
lidar_chm_xr.rio.to_raster(pre_chm_data_path)
```
