## Customize Dates on Time Series Plots

```python
#Importing packages
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import earthpy as et
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
sns.set(font_scale = 1.5, style = 'whitegrid')
```

```python
#Also, import 
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
```

```python
##Working directory
os.chdir(os.path.join(et.io.HOME,'earth-analytics'))
#File path of new data set
file_path = os.path.join('data', 'earthpy-downloads',
                        'colorado-flood','precipitation',
                        '805325-precip-dailysum-2003-2013.csv')
#Accesing to data
precip_2003_2013 = pd.read_csv(file_path, parse_dates = ['DATE'],
                              index_col = ['DATE'],
                              na_values=['999.99'])
```

```python
#Subset
#Period June 1, 2005 - August 31, 2005
precip_june_aug_2005 = precip_2003_2013['2005-06-01':'2005-08-31']
precip_june_aug_2005.head()
```

```python
#Plotting the data
f, ax = plt.subplots(figsize=(10,10))
ax.bar(precip_june_aug_2005.index.values,
          precip_june_aug_2005['DAILY_PRECIP'],
          color = 'purple')
ax.set(xlabel='Date',
      ylabel='Precipitation (inches)',
      title = 'Precipitation\nJune-Aug 2005')
plt.show()
```

```python
#Plotting the data
f, ax = plt.subplots(figsize=(10,10))
ax.bar(precip_june_aug_2005.index.values,
          precip_june_aug_2005['DAILY_PRECIP'],
          color = 'purple')
ax.set(xlabel='Date',
      ylabel='Precipitation (inches)',
      title = 'Precipitation\nJune-Aug 2005')
date_form = DateFormatter('%m-%d')
ax.xaxis.set_major_formatter(date_form)
plt.show()
```

```python
#Plotting the data
f, ax = plt.subplots(figsize=(12,12))
ax.bar(precip_june_aug_2005.index.values,
          precip_june_aug_2005['DAILY_PRECIP'],
          color = 'purple')
ax.set(xlabel='Date',
      ylabel='Precipitation (inches)',
      title = 'Precipitation\nJune-Aug 2005',xlim=['2005-06-01','2005-08-31'])
date_form = DateFormatter('%m-%d')
ax.xaxis.set_major_formatter(date_form)

#Modifying frequency of xaxis
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))
plt.show()
```

```python
#Pratice
#Data set
#Time Period A: 2012-08-01 to 2012-11-01
precip_aug_nov_2012 = precip_2003_2013['2012-08-01':'2012-11-01']
#Time Period B: 2013-08-01 to 2013-11-01
precip_aug_nov_2013 = precip_2003_2013['2013-08-01':'2013-11-01']
```

```python
#PLot
f, (ax1, ax2) = plt.subplots(2,1, figsize = (15,15))
ax1.bar(precip_aug_nov_2013.index.values,
       precip_aug_nov_2013['DAILY_PRECIP'],
       color = 'purple')
ax1.set(xlabel = 'Date\n',
       ylabel='Precipitation (inches)',
       title = 'Daily Total Precipitation\nAug-Oct 2013 for Boulder Creek',
       ylim = [0,10])
ax2.bar(precip_aug_nov_2012.index.values,
       precip_aug_nov_2012['DAILY_PRECIP'],
       color = 'purple')
ax2.set(xlabel = 'Date',
       ylabel='Precipitation (inches)',
       title = 'Daily Total Precipitation\nAug-Oct 2012 for Boulder Creek')
date_form = DateFormatter('%b %d')
ax1.xaxis.set_major_formatter(date_form)
ax2.xaxis.set_major_formatter(date_form)
plt.show()
```
