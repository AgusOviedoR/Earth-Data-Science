```python
#Importing packages
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import earthpy as et
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
sns.set(font_scale = 1.5, style = 'whitegrid')
```

```python
#Working directory
os.chdir(os.path.join(et.io.HOME, 'earth-analytics'))
#File path of new data set
file_path = os.path.join('data', 'earthpy-downloads',
                        'colorado-flood',
                        'precipitation','805325-precip-daily-2003-2013.csv')
#Accessing data
boulder_precip_2003_2013 = pd.read_csv(file_path, parse_dates=['DATE'],
                                      index_col = ['DATE'],
                                      na_values = ['999.99'])
boulder_precip_2003_2013.head()
```

```python
#PLot
f, ax = plt.subplots(figsize=(10,10))

ax.scatter(boulder_precip_2003_2013.index.values,
          boulder_precip_2003_2013['HPCP'],
          color = 'purple')
ax.set(xlabel = 'Date',
      ylabel='Precipitation (inches)',
      title = 'Hourly Precipitation - Boulder Station\n2003-2013')
plt.show()
```

```python
#Resampling to a daily set
precip_daily_2003_2013 = boulder_precip_2003_2013.resample('D').sum()
```

```python
#PLot
f, ax = plt.subplots(figsize=(10,10))

ax.scatter(precip_daily_2003_2013.index.values,
          precip_daily_2003_2013['HPCP'],
          color = 'purple')
ax.set(xlabel = 'Date',
      ylabel='Precipitation (inches)',
      title = 'Daily Precipitation - Boulder Station\n2003-2013')
plt.show()
```

```python
#Resampling to a dmonthlyset
precip_monthly_2003_2013 = boulder_precip_2003_2013.resample('M').sum()
```

```python
#PLot
f, ax = plt.subplots(figsize=(10,10))

ax.scatter(precip_monthly_2003_2013.index.values,
          precip_monthly_2003_2013['HPCP'],
          color = 'purple')
ax.set(xlabel = 'Date',
      ylabel='Precipitation (inches)',
      title = 'Monthly Precipitation - Boulder Station\n2003-2013')
plt.show()
```

```python
#Yearly resampling
precip_yearly_2003_2013 = boulder_precip_2003_2013.resample('Y').sum()
```

```python
#Plot
f, ax = plt.subplots(figsize=(10,10))
ax.scatter(precip_yearly_2003_2013.index.values,
          precip_yearly_2003_2013['HPCP'],
          color = 'purple')
ax.set(xlabel='Year',
      ylabel ='Precipitation (inches)',
      title = 'Yearly Precipitation - Boulder Station\n2003-2013')
plt.show()
```
