```python
#Importing packages
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import earthpy as et
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
sns.set(font_scale = 1.5, style = 'whitegrid')
```

```python
#Working directory
os.chdir(os.path.join(et.io.HOME, 'earth-analytics','data'))

stream_discharge_path = os.path.join('earthpy-downloads', 'colorado-flood',
                                    'discharge','06730200-discharge-daily-1986-2013.csv')
discharge_1986_2013 = pd.read_csv(stream_discharge_path)

discharge_1986_2013
```

```python
#Manage data
disch_1986_2013 = pd.read_csv(stream_discharge_path,parse_dates=['datetime'],
                             index_col = ['datetime'])
f, ax = plt.subplots(figsize = (12,12))
ax.plot(disch_1986_2013.index.values,
      disch_1986_2013['disValue'],
       color = 'purple')
ax.set(xlabel = 'Date',
      ylabel='Discharges',
      title = 'Daily Discharge Boulder\n1986-2013')
plt.show()
```

```python
#Subsetting period during flood
disch_aug_oct_2013 = disch_1986_2013['2013-08-01':'2013-10-31']
f, ax = plt.subplots(figsize =(12,12))
ax.scatter(disch_aug_oct_2013.index.values,
          disch_aug_oct_2013['disValue'], color ='purple')
ax.set(xlabel='Date',
      ylabel='Discharge',title = 'Discharge Values for Boulder Creek\nAugust - October, 2013')
date_form = DateFormatter('%b-%d')
ax.xaxis.set_major_formatter(date_form)
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))
f.autofmt_xdate()
```

```python
#Resampling
#Maximun discharge value for each week
max_disch_810_2013 = disch_aug_oct_2013.resample('W').max()
max_disch_810_2013
```

```python
#Plot
f, ax = plt.subplots(figsize=(12,12))
ax.scatter(max_disch_810_2013.index.values,
          max_disch_810_2013['disValue'],color = 'purple')
ax.set(xlabel='Date',
      ylabel='Discharge',
      title='Weekly Maximum Discharge Values for Boulder Creek\nAugust-Oct 2013')
date_form = DateFormatter('%b-%d')
ax.xaxis.set_major_formatter(date_form)
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))
f.autofmt_xdate()
plt.show()
```

```python
#resampling to a daily maximum
#2003
disch_aug_oct_2003 = disch_1986_2013['2003-08-01':'2003-10-31']
dmax_810_2003 = disch_aug_oct_2003.resample('D').max()
dmax_810_2013 = disch_aug_oct_2013.resample('D').max()

#Plot
f, (ax1,ax2) = plt.subplots(2,1, figsize = (15,15))
ax1.scatter(dmax_810_2003.index.values,
           dmax_810_2003['disValue'], color ='purple')
ax1.set(xlabel='Date',ylabel='Discharge',title='Daily Discharge Values for\nBoulder Creek Aug-Oct 2003')
ax2.scatter(dmax_810_2013.index.values,
           dmax_810_2013['disValue'], color ='purple')
ax2.set(xlabel='Date',ylabel='Discharge',title='Daily Discharge Values for\nBoulder Creek Aug-Oct 2013')
ax1.set_ylim(0,5000)
ax2.set_ylim(0,5000)
ax1.set_xlim(dmax_810_2003.index.values.min(),dmax_810_2003.index.values.max())
ax2.set_xlim(dmax_810_2013.index.values.min(),dmax_810_2013.index.values.max())
date_form = DateFormatter('%b-%d')
ax1.xaxis.set_major_formatter(date_form)
ax2.xaxis.set_major_formatter(date_form)
ax1.xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))
ax2.xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))
f.autofmt_xdate()
f.suptitle('ScatterPlots')
plt.tight_layout()
plt.show()
```
