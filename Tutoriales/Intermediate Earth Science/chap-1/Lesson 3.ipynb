```python
#Importing packages
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import earthpy as et

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

sns.set(font_scale = 1.5, style = 'whitegrid')
```

```python
#Change directory
os.chdir(os.path.join(et.io.HOME, 'earth-analytics'))

#File path to data set
file_path = os.path.join('data', 'earthpy-downloads',
                        'colorado-flood','precipitation',
                        '805325-precip-dailysum-2003-2013.csv')
```

```python
#Data set
boulder_precip_2003_2013 = pd.read_csv(file_path, parse_dates = ['DATE'],
                                      index_col = ['DATE'],
                                      na_values = ['999.99'])
#First view
boulder_precip_2003_2013.head()
```

```python
boulder_precip_2003_2013.describe()
```

```python
#Subset data - Selecting year 2013
boulder_precip_2003_2013[boulder_precip_2003_2013.index.year == 2013].tail()
```

```python
##Subsetting all december data
boulder_precip_2003_2013[boulder_precip_2003_2013.index.month==12].head()
```

```python
#Period subset aug-may 2005
precip_may_aug_2005 = boulder_precip_2003_2013['2005-05-01':'2005-08-01']
precip_may_aug_2005.head()

#Plotting
f, ax = plt.subplots(figsize = (10,10))
ax.bar(precip_may_aug_2005.index.values,
      precip_may_aug_2005['DAILY_PRECIP'],
      color = 'purple')
ax.set(xlabel='Date',
      ylabel = 'Precipitation (inches)',
      title = 'Daily Total Precipitation\nMay-Aug 2005 Boulder Creek')
plt.setp(ax.get_xticklabels(), rotation = 45)
plt.show()
```
