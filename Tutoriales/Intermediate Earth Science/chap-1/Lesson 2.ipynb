```python
##Importing packages
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import earthpy as et

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

sns.set(font_scale = 1.5, style = 'whitegrid')
```

```python
##Boulder Colorado
#Data: Precipitation
file_url = 'https://ndownloader.figshare.com/files/12948515'
os.chdir(os.path.join(et.io.HOME, 'earth-analytics'))
et.data.get_data(url = file_url)
file_path = os.path.join('data', 'earthpy-downloads',
                        'july-2018-temperature-precip.csv')

#Pandas dataframe
boulder_july_2018 = pd.read_csv(file_path)
```

```python
boulder_july_2018.head()
```

```python
boulder_july_2018.info()
```

```python
#Plot dates as str
f, ax = plt.subplots(figsize=(10,10))
ax.plot(boulder_july_2018['date'],
       boulder_july_2018['precip'],
       color = 'purple')
ax.set(xlabel='Date',
      ylabel='Precipitation (inches)',
      title='Daily Precipitation\nBoulder, Colorado July 2018')
plt.show()
```

```python
#Importing data using datetime and  set index to datetime
boulder_july_2018=pd.read_csv(file_path,
                             parse_dates = ['date'],
                             index_col = ['date'], na_values=['-999.00', 'NA', ' '])
boulder_july_2018
```

```python
f, ax = plt.subplots(figsize = (10,10))
ax.scatter(boulder_july_2018.index.values,
          boulder_july_2018['precip'],
          color = 'purple')
ax.set(ylim=[0,0.5],xlabel='Date',
      ylabel='Precipitation (inches)',
      title='Daily Total Precipitation \n Boulder, Colorado in July 2018')
plt.setp(ax.get_xticklabels(),rotation=45)
plt.show()
```
