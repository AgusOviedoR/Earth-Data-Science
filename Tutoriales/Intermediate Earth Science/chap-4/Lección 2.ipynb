##### Raster Data
Manejo de Raster Data utilizando Xarray 

```python
#Importando librerias a utilizar
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import rioxarray as rxr
import earthpy as et

#Configuracion de graficos con seaborn
sns.set(font_scale = 1.5,style = 'white')
```

```python
#Acceso a datos dentro del directorio de descargas
os.chdir(os.path.join(et.io.HOME,'earth-analytics',
                     'data'))
#Direccion a los archivos a utilizar
dem_pre_path = os.path.join('earthpy-downloads','colorado-flood', 'spatial',
                           'boulder-leehill-rd', 'pre-flood',
                           'lidar','pre_DTM.tif')
#Abrir archivo con rioxarray
dtm_pre_arr = rxr.open_rasterio(dem_pre_path)

#Visualizar archivo
dtm_pre_arr
```

Al abrir un raster data utilizando rioxarray se crea un `xarray.DataArray` que guarda:
* raster data en formato array de numpy
* metadata espacial que incluye crs
* cualquier otro metadata (ejemplo: valor del nodata)

```python
#Ver el CRS y la extensión
print('El CRS para estos datos es: ', dtm_pre_arr.rio.crs)
print('Los limites son: ', dtm_pre_arr.rio.bounds())
```

```python
#Ver el no data value
print('El valor del no data es: ', dtm_pre_arr.rio.nodata)
```

```python
#Una vez importados los datos, podemos graficar usando xarray.plot()
dtm_pre_arr.plot()
plt.show()
```

Si nos encontramos con una gráfica sin sentido debemos explorar los nodata existentes en el array. Estos valores nulos afectan negativamente nuestros graficos.
Con un histograma podemos observar la cantidad o existencia de `no data`

```python
#Histograma para observar rango de valores en los datos
dtm_pre_arr.plot.hist(color = 'purple')
plt.show()
```

Tenemos una cantidad de valores menores a 0, y por el valor del no data visto anteriormente,existen en nuestros datos

```python
#Volvemos a acceder a los datos, pero esta vez utilizando el parámetro masked = True
#este indica que debemos ignorar los valores nodata.
# Además, tiene una columna de mas que dificulta la grafica
dtm_pre_arr = rxr.open_rasterio(dem_pre_path, masked = True).squeeze()
print(dtm_pre_arr.shape)
```

```python
#Graficando de vuelta
f, (ax1, ax2) = plt.subplots(2,1, figsize = (20,10))
dtm_pre_arr.plot(cmap = 'Greys_r',
                 ax = ax1)
ax1.set_title('Modelo de Elevación (DEM)\Inundación en Boulder 2013')
ax1.set_axis_off()
dtm_pre_arr.plot.hist(color='purple', bins =20,ax =ax2)
ax2.set_title('Histograma de los valores\nNo data removidos')
plt.show()
```

```python
#Limites del sitio - Shapefile
site_bound_path = os.path.join("earthpy-downloads","colorado-flood",
                               "spatial",
                               "boulder-leehill-rd",
                               "clip-extent.shp")
site_bound_shp = gpd.read_file(site_bound_path)

# Graficando vector data
f, ax = plt.subplots(figsize=(8,4))
site_bound_shp.plot(color='teal',
                    edgecolor='black',
                    ax=ax)
ax.set(title="Site Boundary Layer - Shapefile")
plt.show()
```

```python
#Grafico de los limites y la capa Raster solapada
f, ax = plt.subplots(figsize=(11, 4))

dtm_pre_arr.plot.imshow(cmap="Greys",
                        ax=ax)
site_bound_shp.plot(color='None',
                    edgecolor='teal',
                    linewidth=2,
                    ax=ax,
                    zorder=4)

ax.set(title="Raster Layer with Vector Overlay")
ax.axis('off')
plt.show()
```
