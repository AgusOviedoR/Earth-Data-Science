##### Ejercicios con Raster Data

```python
#Importando librerias a utilizar
import os
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import rioxarray as rxr
import rasterio as rio
from rasterio.plot import plotting_extent

import earthpy as et
import earthpy.plot as ep

#Configuracion de graficos
sns.set(font_scale = 1.5, style = 'white')
```

```python
#Directorio de trabajo
os.chdir(os.path.join(et.io.HOME, 'earth-analytics','data'))

#Direccion archivos: Ejercicio 1
pre_hill_path = os.path.join('earthpy-downloads','colorado-flood',
                            'spatial','boulder-leehill-rd','pre-flood',
                            'lidar','pre_DTM_hill.tif')
#Acceso a archivo .tif
pre_DTM_hill = rxr.open_rasterio(pre_hill_path, masked = True)
#Visualización
pre_DTM_hill
```

```python
#Grafico
f,ax=plt.subplots(figsize =(10,5))
ep.plot_bands(pre_DTM_hill, cmap='Greys_r', ax =ax)
ax.set_title('Modelo Digital del Terreno\nSombreado de las elevaciones')
plt.show()
```

```python
#Archivos dtm_pre_arr
pre_arr_path = os.path.join('earthpy-downloads','colorado-flood', 'spatial',
                           'boulder-leehill-rd', 'pre-flood',
                           'lidar','pre_DTM.tif')
dtm_pre_arr = rxr.open_rasterio(pre_arr_path, masked = True).squeeze()
dtm_pre_arr
```

```python
#Graficos solapados
f, ax = plt.subplots(figsize=(10, 6))
ep.plot_bands(dtm_pre_arr, ax = ax, cmap ='viridis')
ep.plot_bands(pre_DTM_hill, cmap='Greys_r', ax =ax, cbar =False, alpha = 0.5)
ax.set(title="Modelo de Terreno Digital\nElevaciones y pre-inundacion")
ax.set_axis_off()
plt.show()
```

```python
#Limites del sitio - Shapefile
site_bound_path = os.path.join("earthpy-downloads","colorado-flood",
                               "spatial",
                               "boulder-leehill-rd",
                               "clip-extent.shp")
site_bound_shp = gpd.read_file(site_bound_path, masked=True)
# Define relative path to file
dem_pre_path = os.path.join("earthpy-downloads","colorado-flood",
                            "spatial",
                            "boulder-leehill-rd",
                            "pre-flood",
                            "lidar",
                            "pre_DTM.tif")

with rio.open(dem_pre_path) as dem_src:
    lidar_dem_im = dem_src.read(1, masked=True)
    # Create an object called lidar_dem_meta that contains the spatial metadata
    lidar_dem_meta = dem_src.profile
# Open the file using a context manager and get the values as a numpy array with .read()
with rio.open(dem_pre_path) as dem_src:
    dtm_pre_arr2 = dem_src.read(1)

dtm_pre_arr2
with rio.open(dem_pre_path) as dem_src:
    lidar_dem_im = dem_src.read(1, masked=True)
    # Create an object called lidar_dem_plot_ext that contains the spatial metadata
    lidar_dem_plot_ext = plotting_extent(dem_src)

# This plotting extent object will be used below to ensure your data overlay correctly
lidar_dem_plot_ext
```

```python
#Graficos solapados mas los limites del sitio
f, ax = plt.subplots(figsize=(10, 6))
ep.plot_bands(dtm_pre_arr, ax = ax, cmap ='viridis', extent=lidar_dem_plot_ext)
ep.plot_bands(pre_DTM_hill, cmap='Greys_r', ax =ax, cbar =False, alpha = 0.5,extent=lidar_dem_plot_ext)
site_bound_shp.plot(ax =ax, edgecolor = 'orange', color = 'None')
ax.set(title="Modelo de Terreno Digital\nElevaciones y pre-inundacion")
ax.set_axis_off()
plt.show()
```
